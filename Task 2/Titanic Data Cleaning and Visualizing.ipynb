# Import Libraries

```python
# Import Necessary Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Dataset

```python
# Load the dataset
titanic_data = pd.read_csv("train.csv")
```

```python
# Display the first few rows of the dataset
display(titanic_data.head())
```

# Check for Missing Values

```python
# Check for missing values
missing_values = titanic_data.isnull().sum()
```

```python
display(missing_values)
```

# Replace the Missing Values

```python
# Fill missing values
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace = True)

titanic_data['Cabin'].fillna('Unknown', inplace = True)

titanic_data['Embarked'].fillna('Unknown', inplace = True)
```

# Data Visualization

```python
# Distribution of passengers by sex
sns.countplot(x = 'Sex', data = titanic_data)

plt.title('Passenger Distribution by Sex')

plt.show()
```

```python
# Distribution of passengers by class
sns.countplot(x = 'Pclass', data = titanic_data)

plt.title('Passenger Distribution by Class')

plt.show()
```

```python
# Distribution of passengers by age
sns.histplot(titanic_data['Age'], bins = 20)

plt.title('Age Distribution of Passengers')

plt.xlabel('Age')

plt.show()
```

```python
# Survival rate by sex
sns.barplot(x = 'Sex', y = 'Survived', data = titanic_data)

plt.title('Survival Rate by Sex')

plt.show()
```

```python
# Survival rate by class
sns.barplot(x = 'Pclass', y = 'Survived', data = titanic_data)

plt.title('Survival Rate by Class')

plt.show()
```

```python
# Survival rate by age
survived = titanic_data[titanic_data['Survived'] == 1]

not_survived = titanic_data[titanic_data['Survived'] == 0]

sns.histplot(survived['Age'], bins = 20, kde = True, color = 'blue', alpha = 0.5, label = 'Survived')

sns.histplot(not_survived['Age'], bins = 20, kde = True, color = 'red', alpha = 0.5, label = 'Not Survived')

plt.title('Survival by Age')

plt.xlabel('Age')

plt.legend()

plt.show()
```
